```python
data = np.random.randint(-100,100,50).cumsum()
```

```python
data
```

```python
plt.plot(range(50),data,'r')
plt.title('시간별 가격 추이')
plt.ylabel('주식가격')
plt.xlabel('시간(분)')
```

```python
import matplotlib.font_manager as fm
```

```python
fm.fontManager.ttflist[:5]
```

```python
#path가 필요
path = 'C:\\Windows\\Fonts\\malgunsl.ttf'  #해당폰트파일의 경로가 필요함
fontprop = fm.FontProperties(fname=path, size=17)
plt.plot(range(50),data,'r')
plt.title('시간별 가격 추이', fontproperties=fontprop)
plt.xlabel('주식가격', fontproperties=fontprop)
plt.ylabel('시간(분)', fontproperties=fontprop)
```
