```python
f=open('[박경리]_토지1.txt', 'r')
story=f.read()
f.close()
```

```python
print(type(story))
print(story)
```

```python
stopwords_kr = ['있었다', '것이다']
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

def displayWordCloud(data = None, backgroundcolor = 'white', width=800, height=600):
    wordcloud = WordCloud(
                        font_path = 'C:\\Windows\\Fonts\\malgunsl.ttf',
                        stopwords = stopwords_kr,  #이 글자들은 제외시키고 만들겠다.
                        background_color = backgroundcolor,
                        width = width,
                        height = height).generate(data)
    #print(wordcloud.words_)
    plt.figure(figsize = (15,10))
    plt.imshow(wordcloud)
    plt.axis("off")  #차트 축 눈금 안보이게
    plt.show()
```

```python
%time displayWordCloud(story)
```

```python
from soynlp.noun import LRNounExtractor
noun_extractor = LRNounExtractor()
nouns = noun_extractor.train_extract([story]) #list of str like
```

```python
type(nouns)
```

```python
displayWordCloud(' '.join(nouns))  #str로 바꾸기
```
