```python
#lifeExp열을 연도별로 그룹화하여 평균 계산하기

```

```python
df.groupby('year').mean()  #mean()평균
```

```python
df.groupby('year')['lifeExp'].mean()  #lifeExp값만 보고싶을 때
```

```python
#lifeExp, gdpPercap열의 평균값을 연도, 지역별로 그룹화하여 계산
```

```python
multi_group_ver = df.groupby(['year','continent'])[['lifeExp','gdpPercap']].mean()  #값이 두개 이상이면 대괄호로 한번 더 묶어줘야함
```

```python
type(multi_group_ver)
```
